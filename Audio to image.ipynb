```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wavfile
```

# Spectogram

```python
Fs, aud = wavfile.read('harvard.wav')
```

```python
aud = aud[:,0]
```

```python
first = aud[:int(Fs*6)]
powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(first, Fs=Fs)
plt.show()
```

# Sounds

```python
import librosa
import librosa.display
```

```python
sound='harvard.wav'
y,sr=librosa.load(sound)
hw, _ =librosa.effects.trim(y)
librosa.display.waveshow(hw, sr=sr)
```

# Mel Spectogram



```python
import numpy as np

```

```python
S = librosa.feature.melspectrogram(hw, sr=sr, n_fft=2048, hop_length=152, n_mels=128)
S_DB = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S_DB, sr=sr, hop_length=152, x_axis='time', y_axis='mel');
plt.colorbar(format='%+2.0f dB');
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
signal, sr = librosa.load(r"C:\Users\SHANMUKH\NOTEBOOK\Spectogram\harvard.wav")
```

```python
import spafe
```

```python
# Plot the waveform
plt.figure(figsize=(10, 4))
plt.plot(signal)
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.title("Waveform")
plt.tight_layout()
plt.show()
```

```python
from spafe.features.mfcc import mfcc
```

```python
# Extract the MFCCs
mfccs = mfcc(signal)
```

```python
mfccs
```

```python
# Plot the MFCCs as a heatmap
plt.imshow(mfccs, origin='lower', aspect='auto', cmap='hot')
plt.title('MFCCs')
plt.xlabel('Frame')
plt.ylabel('MFCC Coefficients')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
data = pd.read_csv('updated_coughvid.csv')
```

```python
# Separate the target variable from the features
X = data.iloc[:, :-1] # Features
y = data.iloc[:, -1] # Target variable
```

```python
X = np.abs(X)
```

```python
# Apply feature selection using chi-squared test
best_features = SelectKBest(score_func=chi2, k=60)
fit = best_features.fit(X, y)


# Get the scores and p-values for each feature
scores = fit.scores_
p_values = fit.pvalues_
```

```python
# Create a dataframe with the feature names and their scores
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': scores, 'P-Value': p_values})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)

# Select the top features
selected_features = feature_scores.head(60)['Feature'].tolist()

# Subset the original data with the selected features
selected_data = data[selected_features + ['label']]

```

```python
print(selected_features)
```

```python
# Save the selected features to a new CSV file
selected_data.to_csv('selected_features.csv', index=False)
```

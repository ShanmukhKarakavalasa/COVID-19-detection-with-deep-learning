```python
import pandas as pd
import numpy as np
df=pd.read_csv('Coswara_shuffled.csv')

```

```python
df = df.drop('bSpec', axis=1)
df = df.drop('qSpec', axis=1)
df = df.drop('gSpec', axis=1)
df = df.drop('lSpec', axis=1)
```

```python
df=df.drop('file', axis=1)
```

```python
df = df.replace('#NAME?', np.nan)
mean_values = df.mean(axis=0)
df.fillna(mean_values, inplace=True)
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X = pd.DataFrame(X)

# replace zeros with column mean
X = X.replace(0, np.nan)
X = X.fillna(X.mean())

# replace infinity with column mean
X = X.replace(np.inf, np.nan)
X = X.fillna(X.mean())
```

```python
df.update(X.iloc[:, :])
```

```python
# check for missing values
print(df.isna())

# check for NaN values
print(df.isnull())

# check for zero values
print(df.eq(0))

# check for infinity values
print(df.isin([np.inf, -np.inf]))
```
